```python
# Cell 1: Setup
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from src.eda import TrafficViolationsEDA
from src.utils import load_data

# Configure display
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 100)
plt.rcParams['figure.figsize'] = (12, 6)

# Cell 2: Load Data
df = load_data('../data/cleaned/traffic_violations_cleaned.csv')
print(f"Dataset shape: {df.shape}")
df.head()

# Cell 3: Initialize EDA
eda = TrafficViolationsEDA(df)

# Cell 4: Summary Statistics
summary = eda.get_summary_statistics()
for key, value in summary.items():
    print(f"{key}: {value}")

# Cell 5: Top Violations Analysis
violation_df, violation_fig = eda.analyze_top_violations(top_n=15)
plt.show()
violation_df

# Cell 6: Temporal Analysis
temporal_results = eda.analyze_temporal_patterns()

for name, (data, fig) in temporal_results.items():
    print(f"\n{name.upper()} Distribution:")
    plt.show()

# Cell 7: Demographics Analysis
demo_results = eda.analyze_demographics()

for name, (data, fig) in demo_results.items():
    print(f"\n{name.upper()} Analysis:")
    plt.show()

# Cell 8: Vehicle Analysis
vehicle_results = eda.analyze_vehicles()

for name, (data, fig) in vehicle_results.items():
    print(f"\n{name.upper()} Analysis:")
    plt.show()

# Cell 9: Severity Analysis
severity_results = eda.analyze_severity()

for name, (data, fig) in severity_results.items():
    print(f"\n{name.upper()} Analysis:")
    plt.show()

# Cell 10: Geographic Hotspots
hotspot_df, hotspot_fig = eda.analyze_geographic_hotspots(top_n=20)
plt.show()

# Cell 11: Correlation Analysis
corr_matrix, corr_fig = eda.generate_correlation_analysis()
plt.show()

# Cell 12: Generate Report
report = eda.generate_full_report(save_path='../reports/eda_report.txt')
print(report)

# Cell 13: Display All Insights
insights = eda.get_all_insights()
insights
```
